# **Adaptaci√≥ local, canvi clim√†tic i gen√≤mica dels teosints**

El **canvi clim√†tic** est√† transformant r√†pidament les condicions ambientals del planeta. L‚Äôaugment de les temperatures, els canvis en la disponibilitat d‚Äôaigua i les alteracions en l‚Äôestacionalitat afecten totes les esp√®cies, que es veuen obligades a respondre d‚Äôalguna manera. Aquestes respostes poden incloure:

1.   **Migraci√≥ cap a zones m√©s favorables**, despla√ßant-se geogr√†ficament per mantenir-se dins del seu rang √≤ptim de condicions ambientals.
2.   **Plasticitat fenot√≠pica**, √©s a dir, la capacitat d‚Äôun mateix genotip de produir fenotips diferents en funci√≥ de l‚Äôambient, sense que hi hagi canvis gen√®tics. Aquesta resposta pot ser r√†pida i permetre la superviv√®ncia a curt termini davant canvis ambientals sobtats.
3.   **Adaptaci√≥ gen√®tica**, √©s a dir, la selecci√≥ natural d‚Äôaquelles variants gen√®tiques que confereixen una major aptitud en les noves condicions ambientals, augmentant la seva freq√º√®ncia al llarg de les generacions.

L‚Äôestudi de milers de **variants gen√®tiques** distribu√Ødes al llarg del genoma (SNPs, *Single Nucleotide Polymorphisms*) ens permet detectar patrons d‚Äôadaptaci√≥, identificar poblacions vulnerables i predir com els organismes podrien respondre als escenaris clim√†tics futurs.

# **La hip√≤tesi de l'abund√†ncia al centre de la distribuci√≥**

La **hip√≤tesi de l'abund√†ncia al centre de la distribuci√≥** (*central-abundance hypothesis*) proposa que les poblacions d‚Äôuna esp√®cie no s√≥n iguals a tot el seu rang de distribuci√≥. Segons aquesta hip√≤tesi, les caracter√≠stiques gen√®tiques d‚Äôuna poblaci√≥ depenen de com de lluny es troba del centre de la distribuci√≥ de l‚Äôesp√®cie, tant des del punt de vista geogr√†fic com ambiental.

En general:

1.   Les **poblacions centrals** viuen en condicions ambientals m√©s favorables. Aix√≤ fa que tinguin m√©s individus, una diversitat gen√®tica elevada i una pressi√≥ selectiva relativament baixa.
2.   Les **poblacions perif√®riques** es troben als l√≠mits de la distribuci√≥ de l‚Äôesp√®cie, on les condicions ambientals s√≥n m√©s extremes. Aquestes poblacions solen tenir menys diversitat gen√®tica, per√≤ estan sotmeses a una selecci√≥ natural m√©s intensa, que pot afavorir l‚Äôaparici√≥ d‚Äôadaptacions locals.

Aquesta hip√≤tesi ens ajuda a entendre per qu√® les poblacions que viuen en ambients m√©s dif√≠cils poden mostrar empremptes gen√®tiques d‚Äôadaptaci√≥, especialment rellevants en un context de canvi clim√†tic.

# **Diversitat gen√®tica i dist√†ncia al centre de la distribuci√≥**

Una de les prediccions principals de la hip√≤tesi de l‚Äôabund√†ncia al centre de la distribuci√≥ √©s que la diversitat gen√®tica d‚Äôuna poblaci√≥ disminueix a mesura que ens allunyem del centre de la distribuci√≥ geogr√†fica i ambiental de l‚Äôesp√®cie.

En aquest apartat analitzarem si aquesta predicci√≥ es compleix en els teosints, estudiant la relaci√≥ entre la **diversitat gen√®tica** de cada poblaci√≥ (`Hs`), i:
- la seva **dist√†ncia al centre geogr√†fic** (`Dist_geogr√†fica`),
- la seva **dist√†ncia al centre ambiental** (`Dist_ambiental`).

Per fer-ho, representarem gr√†ficament aquestes relacions i calcularem el coeficient de correlaci√≥ (`r`), que ens permet quantificar la for√ßa i el sentit de l‚Äôassociaci√≥ entre dues variables, i la significaci√≥ estad√≠stica (`p`) de la relaci√≥.

<div style="background-color:#f2f6fb; padding:10px; border-radius:6px; margin:10px 0;">

### **Instruccions**

1. Executa el codi de sota per representar la relaci√≥ entre la diversitat gen√®tica i la dist√†ncia geogr√†fica:

</div>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Separar subesp√®cies i eliminar NaNs
parv = df[df["Subesp√®cie"] == "parviglumis"][["Hs", "Dist_geogr√†fica"]].dropna()
mex  = df[df["Subesp√®cie"] == "mexicana"][["Hs", "Dist_geogr√†fica"]].dropna()

fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

# --- parviglumis ---
x = parv["Dist_geogr√†fica"]
y = parv["Hs"]
r, p = pearsonr(x, y)

axes[0].scatter(x, y, color="tab:blue", alpha=0.7)
m, b = np.polyfit(x, y, 1)
axes[0].plot(x, m*x + b, color="tab:blue")
axes[0].set_title("parviglumis")
axes[0].set_xlabel("Dist√†ncia al centre geogr√†fic")
axes[0].set_ylabel("Diversitat gen√®tica (Hs)")
axes[0].text(0.05, 0.95, f"r = {r:.2f}\np = {p:.3f}",
             transform=axes[0].transAxes, va="top")

# --- mexicana ---
x = mex["Dist_geogr√†fica"]
y = mex["Hs"]
r, p = pearsonr(x, y)

axes[1].scatter(x, y, color="tab:orange", alpha=0.7)
m, b = np.polyfit(x, y, 1)
axes[1].plot(x, m*x + b, color="tab:orange")
axes[1].set_title("mexicana")
axes[1].set_xlabel("Dist√†ncia al centre geogr√†fic")
axes[1].text(0.05, 0.95, f"r = {r:.2f}\np = {p:.3f}",
             transform=axes[1].transAxes, va="top")

plt.suptitle("Hs vs dist√†ncia geogr√†fica per subesp√®cie")
plt.tight_layout()
plt.show()
```

# **Simulaci√≥ de l‚Äôevoluci√≥ d‚Äôun al¬∑lel adaptatiu en condicions de canvi clim√†tic**

Fins ara hem analitzat dades gen√®tiques reals de diverses poblacions de teosints i les hem relacionat amb dades geogr√†fiques i ambientals. Per√≤, qu√® pot passar en el futur si les condicions ambientals continuen canviant?

En aquest apartat simularem l‚Äôevoluci√≥ de la freq√º√®ncia d‚Äôun al¬∑lel adaptatiu al llarg del temps, assumint que aquest al¬∑lel confereix un avantatge selectiu en un ambient m√©s c√†lid o m√©s sec, com el que es preveu amb el canvi clim√†tic.

La simulaci√≥ ens permetr√† explorar com influeixen factors com:
- la intensitat de la selecci√≥ natural (`s`),
- i la mida poblacional (`N`),

en la probabilitat que un al¬∑lel adaptatiu augmenti de freq√º√®ncia o es perdi.

### **Model de simulaci√≥**

Assumirem que:
- hi ha dos al¬∑lels: **A** i **a**
- l‚Äôal¬∑lel **A** t√© un avantatge selectiu (`s > 0`),
- la poblaci√≥ t√© una mida constant al llarg del temps (`N`),
- la reproducci√≥ √©s aleat√≤ria (model *Wright‚ÄìFisher*).

<div style="background-color:#f2f6fb; padding:10px; border-radius:6px; margin:10px 0;">

### **Instruccions**

1. El codi que trobar√†s a continuaci√≥ cont√© dues l√≠nies amb els par√†metres que haur√†s de modificar (`valors_s` i `valors_N`). Segueix les instruccions de cada apartat a continuaci√≥, modifica els valors `valors_s` i `valors_N` del bloc de codi, i torna a executar-lo per observar com canvien els resultats.

2. La simulaci√≥ inclou una component d‚Äôatzar (deriva gen√®tica). Aix√≤ fa que els resultats no siguin exactament iguals cada vegada que s‚Äôexecuta el codi, fins i tot mantenint els mateixos par√†metres. Executa la simulaci√≥ diverses vegades amb els mateixos valors per identificar el comportament general o la tend√®ncia m√©s probable.

</div>

```python
import numpy as np
import matplotlib.pyplot as plt

def simular_al_lel_adaptatiu(p0, s, N, generacions):
    p = p0
    freq = [p]
    for _ in range(generacions):
        # Selecci√≥
        p_sel = (p * (1 + s)) / (p * (1 + s) + (1 - p))
        # Deriva gen√®tica
        p = np.random.binomial(2*N, p_sel) / (2*N)
        freq.append(p)
    return freq

# --- PAR√ÄMETRES FIXOS ---
p0 = 0.1
generacions = 100

# --- PAR√ÄMETRES A MODIFICAR ---
valors_s = [0.0, 0.05]        # intensitat de la selecci√≥
valors_N = [100, 500]        # mida poblacional

# --- SIMULACI√ì ---
plt.figure(figsize=(8, 6))

for s in valors_s:
    for N in valors_N:
        freq = simular_al_lel_adaptatiu(p0, s, N, generacions)
        plt.plot(freq, label=f"s = {s}, N = {N}")

# --- GR√ÄFIC ---
plt.xlabel("Generacions")
plt.ylabel("Freq√º√®ncia de l'al¬∑lel adaptatiu (A)")
plt.title("Evoluci√≥ d‚Äôun al¬∑lel adaptatiu sota diferents escenaris")
plt.ylim(0, 1)
plt.legend()
plt.show()
```
