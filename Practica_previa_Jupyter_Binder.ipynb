### **Instruccions generals**

La segona prova pràctica de les OBC 2026 es realitza mitjançant un *Jupyter Notebook*, que combina blocs de text i blocs de codi.

Els **blocs de text** contenen les explicacions teòriques (sense fons), les instruccions de cada apartat (fons blau) i les preguntes que hauràs de respondre (fons groc).

Els **blocs de codi** tenen el fons gris i contenen instruccions de programació que permeten analitzar les dades i generar gràfiques i simulacions.

Per executar un bloc de codi, cal:

1. Fer clic **dins** del bloc de codi.

2. Fer clic al **botó** ▶︎ que apareix al menú superior (o prémer *Shift+Enter*).

3. Pots executar qualsevol bloc de codi tantes vegades com vulguis, fent clic repetidament al botó ▶︎ (o prement repetidament *Shift+Enter*).

En general, no caldrà modificar el codi; només caldrà executar-lo per obtenir els resultats. En alguns apartats, però, les instruccions indicaran explícitament que cal modificar algun paràmetre del codi i tornar a executar el bloc per observar com canvien els resultats.

</div>
